```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("/kaggle/input/sample-sales-data/sales_data_sample.csv", encoding='unicode_escape' )
df.head()
```

```python
sns.barplot( data=df, x='ORDERNUMBER', y='SALES', hue='STATUS', color='r')
plt.title ('sales insight on the orders')
```

```python
df.tail()
```

```python
#know the sales in month it rose 
sns.scatterplot(data=df,x='MONTH_ID', y='SALES', hue='STATUS',color='blue')
plt.title('sales versus month')
```

```python
df.sort_index()
```

```python
df.COUNTRY.value_counts()
```

```python
#top countries with most sales 
agg=df.groupby('COUNTRY') ['SALES'].mean().sort_values(ascending=False)
plt.title('top sales in the contries')
sns.lineplot (agg.head(5))
```

```python
#Bottom countries with least sales 
agg=df.groupby ('COUNTRY')['SALES'].mean().sort_values(ascending=True)
plt.title('countries with least sales')
sns.lineplot(agg.head(5))
```

```python
sns.scatterplot(x='QUANTITYORDERED',y='SALES', hue='DEALSIZE', data=df)
```
