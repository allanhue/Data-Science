```python
iris_data=pd.read_csv("/kaggle/input/datasets/allanyengsnxmwas/iris-dataset-flowers-classification-beginneer/iris data (2).csv")
iris_data.head()
```

```python
# Setosa Species (a) Mean and (b) Median of the Sepal Length
setosa = iris_data[iris_data["Species"] == "setosa"]

# Mean and Median of Sepal Length
mean_sepal_length = setosa["Sepal.Length"].mean()
median_sepal_length = setosa["Sepal.Length"].median()

print("Mean:", mean_sepal_length)
print("Median:", median_sepal_length)

# Check data types 
print(iris_data.dtypes)
```

```python
# Filter species
versicolor = iris_data[iris_data["Species"] == "versicolor"]
virginica = iris_data[iris_data["Species"] == "virginica"]


#  Petal Length
versicolor_petal_length_mean = versicolor["Petal.Length"].mean()
versicolor_petal_length_median = versicolor["Petal.Length"].median()

#  Sepal Width
setosa_sepal_width_mean = setosa["Sepal.Width"].mean()
setosa_sepal_width_median = setosa["Sepal.Width"].median()

#Petal Width
virginica_petal_width_mean = virginica["Petal.Width"].mean()
virginica_petal_width_median = virginica["Petal.Width"].median()

print("Mean:", virginica_petal_width_mean)
print("Median:", virginica_petal_width_median)

print("Mean:", setosa_sepal_width_mean)
print("Median:", setosa_sepal_width_median)
```
